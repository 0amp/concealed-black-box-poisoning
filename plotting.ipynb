```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fracs = [0.01, 0.05, 0.1, 0.5]
gaps = [0.05, 0.16, 0.15, 0.3]

sns.set()
sns.barplot(x=fracs, y=gaps)
plt.xlabel('Fraction of Training Data Affected')
plt.ylabel('Accuracy Gap for Poisoned Concept')
plt.title('Untargeted Attack for GPT-2 on IMDB')

# add a dashed line for roofline
plt.axhline(y=0.4, color='black', linestyle='--', label='Roofline')
plt.legend()
plt.show()
```
